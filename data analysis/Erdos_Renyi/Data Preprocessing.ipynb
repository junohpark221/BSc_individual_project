```python
import pandas as pd
import numpy as np
import csv
```

```python
f = open('er_cas_raw_100.csv', 'rt')
reader = csv.reader(f)

data_list = []
for line in reader:
    data_list.append(line)

f.close()

er_cas_raw_100 = pd.DataFrame(data_list)
er_cas_raw_100 = er_cas_raw_100.replace('', 0)

cur_col = er_cas_raw_100.columns
new_col = ['rep', 'attack_size', 't', 'init_intra_edge_a', 'init_intra_edge_b', 'init_supp_edge', 'fin_intra_edge_a', 'fin_intra_edge_b', 'fin_supp_edge', 'init_mean_deg_a', 'init_mean_deg_b', 'alive_nodes', 'tot_isol_node', 'tot_unsupp_node', 'cas_steps', 'init_larg_comp_a', 'init_larg_comp_b', 'fin_larg_comp_a', 'fin_larg_comp_b']
needed = len(cur_col) - len(new_col)

l = 0
index = 1
while l < needed:
    new_col.append('step%d_isol' % index)
    new_col.append('step%d_unsupp' % index)
    index += 1
    l += 2

    
er_cas_raw_100.columns = new_col
er_cas_raw_100 = er_cas_raw_100.drop(er_cas_raw_100.index[0])
```
